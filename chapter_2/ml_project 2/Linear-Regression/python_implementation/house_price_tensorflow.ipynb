# Predict House Price using Tensorflow

**The framework for defining a neural network as a set of Sequential layers is called keras, so we import that too.**

```python
import numpy as np
import pandas as pd
import tensorflow as tf
```

```python
def house_model():
    '''
    we create the simplest possible neural network,
    It has 1 layer, and that layer has 1 neuron, and the input shape to it is just 3 features.
    sgd as Stochastic Gradient Descent.
    mean_squared_error is our Cost function
    '''
# initialize the model input_shape[3] because we have 3 features
    model = tf.keras.Sequential([tf.keras.layers.Dense(units=1, input_shape=[3])])
    model.compile(optimizer='sgd', loss='mean_squared_error')

    df = pd.read_csv('housing.csv')
    print("First three rows of our data", df.head(3))
    df = (df - np.mean(df)) / np.std(df)
    print("First three rows of our data after normalization", df.head(3))

# just 5 rows for testing
    test_x = df.iloc[:5, :-1] # for testing
    test_y = abs( df.iloc[:5, -1] )# for testing
# all data for training
    xs = np.array( df.iloc[:, :-1]) # retrive features
    ys = np.array(df.iloc[:, -1]) # retrive real y


    '''
    The process of training the neural network, where it 'learns'
    the relationship between the ys and ys is in the model.fit 
    '''
    model.fit(xs, ys, epochs=500)
    return model, test_x, test_y
```

```python
model, test_x, test_y = house_model()
```

```python
print("The testing data\n", test_x)
print("The labels data\n", test_y)
```

```python
print("The weights of the model\n", model.weights)
```
