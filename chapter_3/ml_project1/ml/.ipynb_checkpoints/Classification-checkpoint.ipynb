# Performance Measurements

After we have fit our model we need some way to measure the accuracy of the model.

But we will doing that using the Cross-Validation idea.

## Measuring Accuracy Using Cross-Validation

We will design the cross-validation based on StratifiedKFold which ensure that we have representive ratio of each class in each flod.

For each flod we get the correponding train_index and test_index, then for each flod we use its own model so we create a clone of the clissifer (Constructs a new unfitted estimator), which help each time you run on other flod to work from base.

Also we will use sklearn cross_val_score()

```python
#  internal it split to 80% train 20% test as default
skflods = StratifiedKFold(n_splits=3, random_state=42) 
for train_index, test_index in skflods.split(X_train, y_train_5):
    clone_clf = clone(sgd_clf) # clone the sgd a model with not updated weights
    # split each flod of the data
    X_train_flods = X_train[train_index]
    y_train_flods = y_train_5[train_index]
    
    # The validation set
    X_test_flods = X_train[test_index]
    y_test_flods = y_train_5[test_index]
    
    # Fit new model each time, predict and calculate the accuracy
    clone_clf.fit(X_train_flods, y_train_flods)
    y_pred = clone_clf.predict(X_test_flods)
    n_correct = np.sum(y_pred == y_test_flods)
    print(n_correct/ len(y_pred))
```

```python
# Check how StratifiedKFold split the data for last fold
print(len(y_test_flods))
print(np.sum(y_test_flods == True))
print(np.abs(len(y_test_flods) - np.sum(y_test_flods == True)))

print(len(y_train_flods))
print(np.sum(y_train_flods == True))
print(np.abs(len(y_train_flods) - np.sum(y_train_flods == True)))
```

```python
# instead of our design split Using scikit learn cross_val_score
cross_val_score(sgd_clf, X_train, y_train_5, cv=3, scoring='accuracy')
```

```python
# BaseEstimator to check non-5 digits
class Never5Classifer(BaseEstimator):
    def fit(self, X, y=None):
        return self
    def predict(self, X):
        return np.zeros((len(X), 1), dtype=bool)
    
# over 90% because out of 100% 90% are not 5 digits, 
# so for each time prediction is 90% to be not-5
never_5_clf = Never5Classifer()
cross_val_score(never_5_clf, X_train, y_train_5, cv=3, scoring='accuracy')
```

# Confusion Matrix

Each row represent actuall class, while each column represent a predicted class.

**Non-5-digits** The first row represent the non 5 digits, which is called (the negative class), **53892** of these images were classified as **Non-5-images** correctly which is called **(True Negative)**, on the other hand **1891** of these images are wrongly classified as 5 which is called **(False Positive)**. and this which concerned with the **precision**.

**5-digits** The second row which represent the class 5-images, and its called (the positive class)) which we are looked for, **687** of these images wrongly classified as non 5 images, **(False Negative)**, while the remaining **3530** of these images correctly classified as 5-images. and this which concerned with the **Recall**

# Note !

A lot of information that the matrix has, but some times we looked for the actual class which is the **Positive Class** and this called **precision** 

precision = TP/ TP + FP. (the correct 5-digits only).

So in **precision** we get out the total of correct 5-images the model classifed over total of these correct classifed + where its fail and classifed **Non-5-images** as **5-images**


Since we looked for how model perform on the 5-class and also on the Non-5-class, which actually represent the majority of the digits, so we calculated another thing called **Recall or sensitivity** or **TPR True Positive Ratio**, this the ratio of positive instance that are correctly detected by the classifier.

The name **sensitivity** actually means its more concern on the class 5 where its classifed correct or other classified as **Not-5** but actually they are **5**, in this case its looks for the overall classifed of **digit-5** over the summation of this correct classifed and those are also 5 but they wron classifed as not-5.

Recall = TP/ TP + FN. (the detected 5-digits).

# F1-Score 

When it comes to model performance and compare multiple models together, it's good way to look as one number that help you decide to use this model or not, th **F1-Score**, is combined the **precision and Recall**, the **f1-score** is the harmonic mean of precision and recall, and the harmonic means give more weight to low values, so if one of these **precision or Recall** is low the overall of **f1-score** will be low so we try to make a trade-off between them.

# The ROC Curve

The **Reciver Operating Characters** curve is another tools used with binary classifiers. It as **precision and Recall**, but instead of plot **Precision** Vs **Recall**. it plot **True Positive** another name for **Recall** which is the percentage of correctly classified **5-digit** (detected 5s), aganist the FPR(False Positive Rate) which is the ratio of wrongly classifed as 5s but actually they are not **5-digits**.


To plot ** ROC Curve** then we need to get **FPR** and **TPR**

```python
fpr, tpr, thresholds = roc_curve(y_train_5, y_scores)
```

```python
def plot_roc_curve(fpr, tpr, label=None):
    plt.plot(fpr, tpr, label=label)
    plt.plot([0, 1], [0,1], 'k--') # Diagonal of random classifier
    plt.xlabel("FPR")
    plt.ylabel("Recall-TPR")
```

```python
plot_roc_curve(fpr, tpr)
```

# Multiclass Classifcation

Now let's go away with detect more than just **5-digit**.

Binary Classifier distinguish between two classes, but its not 2-class we have 1 of 10 digits to predit, so its called **Multiple Classifer or Multinomial Classifer**.

Models like SGD Classifier, Random Forest, Naive Bayes are capable of handle multiple classes and Logistic Regression and SVM are strictly binary classifiers.

There are Two ways of MultiClass Classifer:

- OVR or called OVA (One-versue-Rest) or called (One-versue-All).

In this strategy you predict **10-class score** for each image you need to classify, then you choose the maximium score of these classes, the way its work is that like we have with **5-digit** veruse the **other digits**, so its run **10-classifier** one for each digit from [0-9].

- OVO (One-Versue-One).

In this strategy it train a binary classifier for each pair of classes so (0s and 1s), (0s and 2s) and so on to (0s and 9s), then it move to (1s and 0s), (1s and 2s) and so on to (1s and 9s) to last class. So it run in this case 45 binary classifer, **N * (N-1) /2**, N for number of classes. So you will run 45-classifier when you need to classify one image, then you choose also which best score of them. This way have the advantage of train on part of the dataset, just the pairs you deal with now like(0s and 1s), and models like SVC have to work with this prefect because it poorly with the scale of data (large data), so like this 45-classifer is run faster on small parts of the whole data than using all data in case you decide to use OVR with SVC.

Scikit-Learn detects when you try to run **Binary Classifer Model** like **SVM or Logistic Regression** on multiple classification problem, and it run OVR or OVO depends on your model is of those who stricly **Binary Classifer** or **Multiple Classifer**.

So one of the ways w should consider is to analyze the closed loops in the image or make some process to make it appear to the model and take care of the images contain loops because it has very similar representation, maybe complex model can help or other ways.

```python
def plot_digits(instances, images_per_row=10, **options):
    size = 28
    images_per_row = min(len(instances), images_per_row)
    images = [instance.reshape(size,size) for instance in instances]
    n_rows = (len(instances) - 1) // images_per_row + 1
    row_images = []
    n_empty = n_rows * images_per_row - len(instances)
    images.append(np.zeros((size, size * n_empty)))
    for row in range(n_rows):
        rimages = images[row * images_per_row : (row + 1) * images_per_row]
        row_images.append(np.concatenate(rimages, axis=1))
    image = np.concatenate(row_images, axis=0)
    plt.imshow(image, cmap = matplotlib.cm.binary, **options)
    plt.axis("off")
```

```python
cl_a, cl_b = 3,5

# How many images are 3 and predicited as 3
X_aa = X_train[(y_train == cl_a) & (y_train_pred == cl_a)]

# How many images are 3 and predicited as 5
X_ab = X_train[(y_train == cl_a) & (y_train_pred == cl_b)]

# How many images are 5 and predicited as 5
X_bb = X_train[(y_train == cl_b) & (y_train_pred == cl_b)]

# How many images are 5 and predicited as 3
X_ba = X_train[(y_train == cl_b) & (y_train_pred == cl_a)]


plt.figure(figsize=(8,8))
plt.subplot(221); 
plot_digits(X_aa[:25], images_per_row=5)
plt.subplot(222); 
plot_digits(X_ab[:25], images_per_row=5)
plt.subplot(223); 
plot_digits(X_bb[:25], images_per_row=5)
plt.subplot(224); 
plot_digits(X_ba[:25], images_per_row=5)
```

# Q-2

Write a function that can shift an MNIST image in any direction (left, right, up, or down) by onepixel.5 Then, for each image in the training set, create four shifted copies (one per direction) and addthem to the training set. Finally, train your best model on this expanded training set and measure itsaccuracy on the test set. You should observe that your model performs even better now! Thistechnique of artificially growing the training set is called data augmentation or training setexpansion.

```python
def shift_image(image, dx, dy):
    image = image.reshape((28, 28))
    shifted_image = shift(image, [dy, dx], cval=0, mode="constant")
    return shifted_image.reshape([-1])
```

```python
image = X_train[1000] # Take one image
shifted_image_down = shift_image(image, 0, 5)
shifted_image_left = shift_image(image, -5, 0)

plt.figure(figsize=(12,3))
plt.subplot(131)
plt.title("Original", fontsize=14)
plt.imshow(image.reshape(28, 28), interpolation="nearest", cmap="Greys")
plt.subplot(132)
plt.title("Shifted down", fontsize=14)
plt.imshow(shifted_image_down.reshape(28, 28), interpolation="nearest", cmap="Greys")
plt.subplot(133)
plt.title("Shifted left", fontsize=14)
plt.imshow(shifted_image_left.reshape(28, 28), interpolation="nearest", cmap="Greys")
```

```python
X_train_augmented = [image for image in X_train]
y_train_augmented = [label for label in y_train]
```

```python
# right left bottom top 
for dx, dy in ((1, 0), (-1, 0), (0, 1), (0, -1)):
    for image, label in zip(X_train, y_train):
        X_train_augmented.append(shift_image(image, dx, dy)) # append shifted images
        y_train_augmented.append(label)

X_train_augmented = np.array(X_train_augmented)
y_train_augmented = np.array(y_train_augmented)
```
