```python
'''
    Here we assign our features to variables X and the actual out put to y
    np.array to handle the returned value i get from df file to be array
    which help me make a numpy operation on the data
'''
X = np.array(df.iloc[:,:len(columns)-1]) 
y = np.array(df.iloc[:,len(columns)-1:len(columns)])
print("First five rows of features\n\n", X[0:5,:])
print("\n\nFirst five rows of actual output \n\n", y[0:5])
```

```python
'''
    first we make a features normalization
    second extend X to have x0 =1 for each training examples
    create random values for theats to be 3 as 3 features of x
'''
X = (X - np.mean(X))/ np.std(X)
X = np.column_stack([np.ones((X.shape[0],1)), X])
theats = np.zeros((X.shape[1],1))
print("First five rows of features X after normalization \n\n", X[0:5,:])
print("\n\nTheat Values\n\n",theats)
```

```python
def sigmoid(z):
    '''
        Sigmoid function is used with classification
        used to get all values between 0 & 1
        so h_x with classification returned as 0>=h_x<=1
        after I am display what is beside the values returned from sigmoid
        which if z is positive then our sigmoid will return values near ro 1
        if z is negative then it will return value near to 0
        if z = 0 then it will return .5 becaue np.exp(-0) = 1 and 1/(1+1) =.5
    '''
    return 1 / (1+np.exp(-z))
```

```python
'''
    you will notice that when the value of array 0 to 20
    sigmoid is start from .5 to near of 1 on y axis 
    before that when its negative values sigmoid is < .5
    that is because when z is negative 1/(1+ np.exp(-(-z))) = 1/(1+bigvalues) so it near to 0
'''
tes_sigmoid = np.arange(-20, 20)
tes_sigmoid
fig, ax = plt.subplots(figsize=(10,7))
ax.plot(tes_sigmoid, sigmoid(tes_sigmoid),'r--', linewidth=1, marker='d', markersize=13)
plt.xlabel("tes_sigmoid")
plt.ylabel("Apply Sigmoid")
plt.legend(["Sigmoid Function"])
plt.title("Testing Sigmoid Function")
```

```python
def cost_function(X, y, theats, m):
    '''
        here is implementation of cost function with vectorized version
        you can see steps of this in the file uploaded for basic level example 
        called "Hello Logistic Regression" 
    '''
    h_x = sigmoid(np.dot(X,theats))
    part_1 = np.dot(y.T, np.log(h_x))
    part_2 = np.dot((1-y).T , np.log(1 -h_x))
    J = (-1/m) * (part_1+part_2)
    return J
```

```python
cost_error = cost_function(X, y, theats, X.shape[0])
cost_error
```

```python
def gradient_descent(X, y, theats, m, alpha):
    '''
        also here is implementation of gradient descent with vectorized version
        you can see steps of this in the file uploaded for basic level example 
        called "Hello Logistic Regression" 
    '''
    h_x = sigmoid(np.dot(X,theats))
    grad  = theats - (alpha/m) * np.dot(X.T, (h_x - y))
    return grad
```

```python
theats =  gradient_descent(X,y, theats, X.shape[0], .7)
theats
```

```python
for i in range(100):
    '''
        here is we iterate for 100 iteration to minimize the cost function
        and return the parameters theats that minimize your cost function
    '''
    cost_error = cost_function(X, y, theats, X.shape[0])
    theats = gradient_descent(X,y, theats, X.shape[0], .7)
    print('\nCost function is \n\n',cost_error)
    print('\nNew thetas is \n\n',theats)
    
```

```python
cost_error = cost_function(X, y, theats, X.shape[0])
print("Now the cost error is \n\n", cost_error)
```

```python
def predictive_values(theats, X):
    '''
        here is we trying to predict the values using our training examples
        and using the theats we get after 100 iteration
        before of that we normalize X again 
        then we assign 1 for values >=.5 and assign 0 otherwise
    '''
    X =  (X- np.mean(X))/ np.std(X)
    h_x = sigmoid(np.dot(X, theats))
    h_x = (h_x>=.5)+0
    return h_x
```

```python
predictions = predictive_values(theats,X)
correct = [1 if ((a == 1 and b == 1) or (a == 0 and b == 0)) else 0 for (a, b) in zip(predictions, y)]
accuracy = (sum(map(int, correct)) % len(correct))
print ('accuracy = {0}%'.format(accuracy))
```

```python
error_between_predicted_and_actual = np.sum(np.abs(predictions-y))
error_between_predicted_and_actual
```
