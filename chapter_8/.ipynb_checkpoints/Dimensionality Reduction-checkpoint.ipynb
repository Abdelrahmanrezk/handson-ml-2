# Dimensionality Reduction

Most of machine learning problem contain a lot of features like images, and this require hard computing resource beside time consuming, not just that its even hard to find a good solution because with higher dimension risk of overfitting being large.

But with all these features some of them are noise like border in images, stop words and others in text, even the orginal data lose some of the features maybe cause miss understanding in some cases, but it help us to make the process fast and in other words can help get better result than in high dimension.

Note just helping us from the model way it help in the Data Visualization as we leave in 3d world and get better intuition when it in 2-d space.

```python
import numpy as np
```

```python
def distance_between_vectors(n_dimension):
    '''
    Random function give you numbers in range 0-1
    Imagine you have 1000 instance !
    '''
    for d in range(n_dimension):
        vec1 = np.random.rand(d) # number of d (features in vector 1)
        vec2 = np.random.rand(d) # number of d (features in vector 2)
        vec3 = np.random.rand(d) # number of d (features in vector 3)
        result = np.linalg.norm(vec1-vec2) # euclidean distance
        result2 = np.linalg.norm(vec1-vec3) # euclidean distance
        print(result)
        print(result2)
    return True
```

```python
distance_between_vectors(100)
```

# Note !

- Distance between vector increase while the number of features increase.
- Compare vec1 with vec2 and with vec3 you will find most training instances are likely to be far away from each other means that a new instance will likely be far away from any training instance , making predictions much less reliable than in lower dimension.

So we need to reduce the **Curse of Dimensions**
