## Import libraries

```python
import numpy as np
import pandas as pd
from sklearn.model_selection import ShuffleSplit
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LinearRegression
from sklearn.metrics import mean_squared_error
# Import supplementary visualizations code visuals.py
import matplotlib.pyplot as plt
# Pretty display for notebooks
%matplotlib inline
```

```python
# Load the Boston housing dataset
data = pd.read_csv('housing.csv')
prices = data['MEDV'] # Target variable
features = data.drop('MEDV', axis = 1) # feature data

print("Our Data points vs our columns", data.shape)
```

## Statistics of data

```python
data.describe()
```

```python
plt.style.use(['ggplot','dark_background'])
plt.figure(1, figsize=(35, 6))
'''
pos is a three digit integer, where the first digit is the number of rows,
the second the number of columns, and the third the index of the subplot.
https://matplotlib.org/api/_as_gen/matplotlib.pyplot.subplot.html
'''
# graph 1 at index 1
plt.subplot(141) # 161, 1 means take the all row, 6 take 6 from the width, 1 start at index 1
plt.scatter(features['RM'], prices, s=100, color='green', marker='P')
fit = np.polyfit(features['RM'], prices, 1) # fit weights to data
plt.plot(features['RM'], features['RM'] * fit[0] + fit[1], lw=2, color='red')
plt.xlabel('rooms')
plt.ylabel('price')

# graph 2 at index 2
plt.subplot(142)
plt.scatter(features['LSTAT'],prices, s = 100, marker='x',color='#876543')
fit = np.polyfit(features['LSTAT'], prices, 1)
plt.plot(features['LSTAT'], features['LSTAT'] * fit[0] + fit[1], lw=2, color='yellow')
plt.xlabel('working poor ratio')
plt.ylabel('price')

# graph 3 at index 3
plt.subplot(143)
plt.scatter(features['PTRATIO'],prices, s = 100, marker='D', color='#ffff00')
fit = np.polyfit(features['PTRATIO'], prices, 1)
plt.plot(features['PTRATIO'], features['PTRATIO'] * fit[0] + fit[1], lw=2, color='green')
plt.xlabel('students to teacher ratio')
plt.ylabel('price')
```

###  Feature Observation

- **RM** is the average number of rooms among homes in the neighborhood.
- **LSTAT** is the percentage of homeowners in the neighborhood considered "lower class" (working poor).
- **PTRATIO** is the ratio of students to teachers in primary and secondary schools in the neighborhood.



```python
# the train_test_split take the data as array so using numby to convert it to np array
X = np.array(features)
y = np.array(prices)
# its helpful to make features normalization
X = (X - np.mean(X)) / np.std(X)
y = (y - np.mean(y)) / np.std(y)
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.2)
```

```python
LRModel = LinearRegression()
```

```python
LRModel.fit(X_train,y_train)
```

```python
print("The Weights of LRModel is", LRModel.coef_)
```

```python
print("The line intercept y in", LRModel.intercept_)
```

```python
print("The Accuracy score of our Model is:", LRModel.score(X_train,y_train))
```

```python
predict = LRModel.predict(X_train)
```

```python
print("The Accuracy score of our Model in training data is:", mean_squared_error(predict,y_train))
```

```python
predict = LRModel.predict(X_test)
```

```python
print("The Accuracy score of our Model in testing data is:", mean_squared_error(predict,y_test))
```
